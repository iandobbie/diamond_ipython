```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def wav2RGB(wavelength):
    w = int(wavelength)

    # colour
    if w >= 380 and w < 440:
        R = -(w - 440.) / (440. - 350.)
        G = 0.0
        B = 1.0
    elif w >= 440 and w < 490:
        R = 0.0
        G = (w - 440.) / (490. - 440.)
        B = 1.0
    elif w >= 490 and w < 510:
        R = 0.0
        G = 1.0
        B = -(w - 510.) / (510. - 490.)
    elif w >= 510 and w < 580:
        R = (w - 510.) / (580. - 510.)
        G = 1.0
        B = 0.0
    elif w >= 580 and w < 645:
        R = 1.0
        G = -(w - 645.) / (645. - 580.)
        B = 0.0
    elif w >= 645 and w <= 780:
        R = 1.0
        G = 0.0
        B = 0.0
    else:
        R = 0.0
        G = 0.0
        B = 0.0

    # intensity correction
    if w >= 380 and w < 420:
        SSS = 0.3 + 0.7*(w - 350) / (420 - 350)
    elif w >= 420 and w <= 700:
        SSS = 1.0
    elif w > 700 and w <= 780:
        SSS = 0.3 + 0.7*(780 - w) / (780 - 700)
    else:
        SSS = 0.0
    #SSS *= 255

    #return [int(SSS*R), int(SSS*G), int(SSS*B)]
    return [SSS*R, SSS*G, SSS*B]
```

```python
datapath = 'fluorochromes'
datasets = {}

for fname in os.listdir(datapath):
    dname = fname.strip('.csv')
    
    wavelength = []
    excitation = []
    emission = []
    
    with open(os.path.join(datapath, fname)) as f:
        for line in f:
            try:
                c1, c2, c3 = [float(i) for i in line.strip('\n').split(',')]
            except:
                pass
            else:
                wavelength.append(c1)
                excitation.append(c2)
                emission.append(c3)

    datasets.update({dname: {
        'wavelength': wavelength,
        'excitation': excitation,
        'emission': emission,}})
    
    emission_peak = wavelength[max(enumerate(emission), key=lambda x: x[1])[0]]
    datasets[dname].update({'colour': tuple(wav2RGB(emission_peak)),
                            'emission_peak': emission_peak})
    
```

```python
filters = {'FITC': (485, 20),
           'TRITC': (535, 36),
           }
           
fig = plt.figure(figsize=(16,12), dpi=100)
for dname, data in datasets.iteritems():
    plt.plot(data['wavelength'], data['emission'], color=data['colour'], label=dname)
plt.title('emission')
plt.legend(loc=4)
ax = fig.gca()

baseline = -5
for k, v in filters.iteritems():
    mid = v[0]
    low = v[0] - v[1]/2
    high = v[0] + v[1]/2
    rgb = wav2RGB(mid)
    ax.annotate('', 
            xy=(low, baseline-1), xycoords='data',
            xytext=(high, baseline-1), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                            connectionstyle="arc3"),
            color=rg
            )
    ax.annotate(k, xy=(mid, baseline), xytext=(mid, baseline), ha='center')
    ax.vlines(low, *ax.get_ylim(), color=rgb)
    ax.vlines(high, *ax.get_ylim(), color=rgb)
ax.get_xlim()
```
